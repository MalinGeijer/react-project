```python
# 1. Databehandling och analys
# Läs in och analysera ett dataset.
# Studenten har själv valt dataset, definierat problemställning
# Utför datarensning (hantera saknade värden, felaktiga typer, m.m.).

from tensorflow.keras.datasets import mnist
import numpy as np

# Ladda MNIST-datasetet
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Kontroll av ursprunglig form
x_train.shape, x_test.shape

# Kontroll av min och max värden
x_train.min(), x_train.max()
print(f"Pixel min value: {x_train.min()}, Pixel max value: {x_train.max()}")
print(" 0 -> svart, 255 -> vitt")
```

```python
# Normalisera pixelvärden från 0–255 till 0–1 -> all input ska konkurrera på samma villkor och snabba upp beräkningarna
# Konvertera till float32 för att inte tappa precision vid division
x_train_norm = x_train.astype("float32") / 255.0
x_test_norm = x_test.astype("float32") / 255.0
x_train_norm.min(), x_train_norm.max()
print(f"Normalized pixel min value: {x_train_norm.min()}, Normalized pixel max value: {x_train_norm.max()}")
```

```python
# Gör om 28x28 bilderna till vektorer med 784 pixlar. Alltså pixlar i en rad.
print("x_train shape:", x_train.shape)
n_train_samples = x_train_norm.shape[0]
height = x_train_norm.shape[1]
width = x_train_norm.shape[2]

print("x_test shape:", x_test.shape)
n_test_samples = x_test_norm.shape[0]
height2 = x_test_norm.shape[1]
width2 = x_test_norm.shape[2]

nof_pixels = height * width
print(f"Nof pixels per image: {nof_pixels}")

x_train_reshaped = x_train_norm.reshape(n_train_samples, nof_pixels)
x_test_reshaped = x_test_norm.reshape(n_test_samples, nof_pixels)


print("Each row is now a flattened image with 784 pixels.")

print("Training data shape:", x_train_reshaped.shape)
print("Test data shape:", x_test_reshaped.shape)
print("Labels shape:", y_train.shape, y_test.shape)
```
